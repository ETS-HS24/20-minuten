```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
directory_path = '../data/processed'
```

# Topic matching by translation using LSA

```python
df = pd.read_csv(f'{directory_path}/lsa_matches_by_translation.csv')
df.drop('Unnamed: 0', axis='columns', inplace=True)
# df = df.drop_duplicates()
df_german = df['german'].value_counts()
df_french = df['french'].value_counts()
print(df_german)
print(df_french)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5))
df_german.plot(title="Topic matching by translation using LSA - German", legend=True, ax=ax1, kind='bar')
df_french.plot(title="Topic matching by translation using LSA - French", legend=True, ax=ax2, kind='bar',
               color='orange')
plt.tight_layout()
```
